```python
import pandas as pd
import altair as alt
```

```python
hpi_df = pd.read_csv('hpi_data_countries.tsv', sep='\t')
```

```python
hpi_df.head()
```

```python
hpi_df['Region'].value_counts()
```

```python
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
).interactive()
```

```python
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color = 'Region'
).interactive()
```

```python
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color = 'Region',
    tooltip = ['Country','Region','Life Expectancy (years)','Wellbeing (0-10)','Ecological Footprint (gha/capita)']
).interactive()   
```

```python
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color = 'Region',
    tooltip = ['Country','Region','Life Expectancy (years)','Wellbeing (0-10)','Ecological Footprint (gha/capita)']
).interactive()
```

```python
selected_area = alt.selection_interval()
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color = 'Region'
).add_selection(selected_area)
```

```python
selected_area = alt.selection_interval()
alt.Chart(hpi_df).mark_circle().encode(
    x='Wellbeing (0-10):Q',
    y='Happy Planet Index:Q',
    color = alt.condition(selected_area, 'Region:N', alt.value('lightgray'))
).add_selection(selected_area)
```

```python
alt.Chart(hpi_df).mark_bar().encode(
x='Region:N',
y = 'Happy Planet Index:Q'
)
```

```python
alt.Chart(hpi_df).mark_bar().encode(
x='Region:N',
y = 'Happy Planet Index:Q'
).properties(width = 1000)
```
